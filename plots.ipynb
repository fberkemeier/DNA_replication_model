```python
# Example usage
cell_line = "H1"
chr_number = 6
data_type = "error"
threshold = 10**2.8
nmax = 10
nmax2 = 100

error = load_function(cell_line, chr_number, data_type)
local_max0 = argrelextrema(error, np.greater)[0]

local_max = [value for value in local_max0 if error[value] > threshold]
local_max_list = [np.array(range(max(0,maxi - nmax), min(chr_lengths[chr_number-1],maxi + nmax)+1)) for maxi in local_max]
local_max_list2 = [np.array(range(max(0,maxi - nmax2), min(chr_lengths[chr_number-1],maxi + nmax2)+1)) for maxi in local_max]

positions = [[item for sublist in local_max_list2 for item in sublist],[item for sublist in local_max_list for item in sublist]]

generate_plot(
    load_type1="time_data", load_type2=data_type,
    cell_line=cell_line, chr_numbers=[chr_number], chr_number_fragile_sites=[chr_number],
    show_telomeres=False, show_centromeres = False,
    show_fragile_sites=False, site_letters=['A'],
    custom_positions=positions, custom_labels=["High errors", "Peaks"],
    xmin=1e-12, xmax=1e6,
    ymin=0, ymax=570,
    log_x=True, log_y=False, invertyQ=True,
    use_density=False,
    dpi=300,
    map_to_white=False,
    title = cell_line,
    saveQ=False
)
```
